Preparing Histological data

```python
col_metadata.histological_type_other.unique()
```

```python
cancer_type = col_metadata.histological_type_other
print(cancer_type.unique)
```

```python
def classify_cancer_type(x):
    x = str(x)
    if 'adenocarcinoma' in x:
        return 'adenocarcinoma'
    elif 'neuroendocrine' in x:
        return 'neuroendocrine'
    else: 
        return 'unknown'
cancer_type = cancer_type.map(lambda x: classify_cancer_type(x))
cancer_type.value_counts().plot(kind='bar')
```

```python
clean_data_matrix = clean_data_matrix.transpose()
clean_data_matrix.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(clean_data_matrix)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_reduced = pca.fit_transform(X_std)
pca_df = pd.DataFrame(data=X_reduced, columns=['PC1','PC2'])
pca_df.head(5)
```

```python
fig = plt.figure(1, figsize=(10,6))
sns.scatterplot(X_reduced[:, 0],  X_reduced[:, 1])
```

### 3.2 Histological Data Overlay

```python
fig = plt.figure(1, figsize=(10,6))
sns.scatterplot(X_reduced[:, 0],  X_reduced[:, 1],hue=cancer_type)
```

>The neuroendocrine tumors clearly separable from the adenocarcinoma tumors

```python
pca_std = PCA()
X_new = pca_std.fit(X_std)
percent_variance=X_new.explained_variance_ratio_*100

plt.figure()
plt.plot(np.cumsum(pca_std.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Cumulative Explained Variance')
plt.show()
```

```python
explained_var = round(np.cumsum(pca_std.explained_variance_ratio_)[50]*100, 2)
print(f"Around {explained_var}% of variance is explained by the Fisrt 50 principle component ");
```

>Looking at the variance of the PCA. Around ***84.13%*** of variance is explained by the fisrt ***50*** principle components.

## 5. Unsupervised analysis

### 5.1 K-Means Clustering 

```python
from sklearn.cluster import KMeans

clusters = []

for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(X_reduced)
    clusters.append(km.inertia_)
    
fig, ax = plt.subplots()
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Searching for Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')

# Annotate arrow
ax.annotate('Possible Elbow Point', xy=(3, 280000), xytext=(1, 100000), xycoords='data',          
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))

ax.annotate('Possible Elbow Point', xy=(4, 200000), xytext=(3, 300000), xycoords='data',          
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))

ax.annotate('Possible Elbow Point', xy=(5, 150000), xytext=(6, 400000), xycoords='data',          
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))


plt.show()
```

Elbows observed at 3,4,5

```python
km3 = KMeans(n_clusters=3).fit(X_reduced)

fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(221)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=km3.labels_, 
                palette=sns.color_palette('hls', 3))
ax.set_title('KMeans with 3 Clusters')

ax = fig.add_subplot(222)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1],
                hue=cancer_type)
ax.set_title('KMeans with 3 Clusters')

ax = fig.add_subplot(223)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1],
                hue=col_metadata.sample_type)
ax.set_title('KMeans with 3 Clusters')

ax = fig.add_subplot(224)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1],
                hue=col_metadata.mRNAseq_cluster)
ax.set_title('KMeans with 3 Clusters')

plt.show()
```

```python
km4 = KMeans(n_clusters=4).fit(X_reduced)

fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(221)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=km4.labels_, 
                palette=sns.color_palette('hls', 4))
ax.set_title('KMeans with 3 Clusters')

ax = fig.add_subplot(222)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1],
                hue=cancer_type)
ax.set_title('KMeans with 3 Clusters')

ax = fig.add_subplot(223)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1],
                hue=col_metadata.sample_type)
ax.set_title('KMeans with 3 Clusters')

ax = fig.add_subplot(224)
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1],
                hue=col_metadata.mRNAseq_cluster)
ax.set_title('KMeans with 3 Clusters')

plt.show()
```

### 5.2 DBSCAN Clustering 

Trying out parameters for DBSCAN clustering

```python
from sklearn.cluster import DBSCAN 

db = DBSCAN(eps=40, min_samples=5).fit(X_reduced)

plt.figure(figsize=(12, 8))
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=db.labels_, 
                palette=sns.color_palette('hls', np.unique(db.labels_).shape[0]))
plt.title('DBSCAN with epsilon 11, min samples 6')
plt.show()
```

### 5.3 Comparing Clustering Algorithms

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy 
from scipy.spatial import distance_matrix 
from sklearn.cluster import DBSCAN 
from sklearn.cluster import MeanShift, estimate_bandwidth

fig = plt.figure(figsize=(20,15))

##### KMeans #####
ax = fig.add_subplot(221)

km5 = KMeans(n_clusters=3).fit(X_reduced)
labels = km5.labels_
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=labels, style=labels,
                palette=sns.color_palette('hls', 3), s=60, ax=ax)
ax.set_title('KMeans with 3 Clusters')


##### Agglomerative Clustering #####
ax = fig.add_subplot(222)

agglom = AgglomerativeClustering(n_clusters=3, linkage='average').fit(X_reduced)
labels = agglom.labels_
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=labels, style=labels,
                palette=sns.color_palette('hls', 3), s=60, ax=ax)
ax.set_title('Agglomerative with 3 Clusters')


##### DBSCAN #####
ax = fig.add_subplot(223)

db = DBSCAN(eps=40, min_samples=5).fit(X_reduced)
labels = db.labels_
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=labels, style=labels, s=60,
                palette=sns.color_palette('hls', np.unique(db.labels_).shape[0]), ax=ax)
ax.set_title('DBSCAN with epsilon 40, min samples 5')


##### MEAN SHIFT #####
ax = fig.add_subplot(224)

bandwidth = estimate_bandwidth(X_reduced, quantile=0.5)
ms = MeanShift(bandwidth).fit(X_reduced)
labels = ms.labels_
sns.scatterplot(X_reduced[:, 0], X_reduced[:, 1], hue=labels, s=60,
                palette=sns.color_palette('hls', np.unique(ms.labels_).shape[0]), ax=ax)
ax.set_title('MeanShift')

plt.tight_layout()
plt.show()
```
